```python
#Â¿CuÃ¡l es el porcentaje de valores nulos en cada columna?

nulos=df.isnull().sum()
porcentaje=nulos/conteo * 100
porcentaje
```

```python
#Sacad los valores Ãºnicos y su frecuencia de las variables numÃ©ricas. Almacenad los resultados en un diccionario. Uno para cada variable.
dict_valores = {}  #Definimos un diccionario vacÃ­o para poder aÃ±adir valores y key.

for col in df:
   unicos=df[col]. unique ()
   unicos_cont=df.value_counts(col)
   for k, v in zip(unicos, unicos_cont): 
      dict_valores[k] = v #Para cada key y value en el bucle for asignamos un valor a cada uno a partir del zip de las dos listas.
print(dict_valores)
```

```python
#Cambiad el nombre de las columnas para que todas estÃ©n en minÃºsculas

columns_minusculas = {col :  col.lower() for col in df.columns}
columns_minusculas
```

```python
df.rename(columns=columns_minusculas,inplace=True)
```

```python
df.head(1)
```

```python
# y reemplazad los espacios por _.

espacios_columns={col:col.strip() for col in df.columns}
espacios_columns
```

```python
df.replace(to_replace= df.columns,value=espacios_columns,inplace=True)

df.head(1)
```

```python
#Eliminad las columnas de Unnamed: 22 y Unnamed: 23, Case Number.1, Case Number.2, original order, Investigator or Source, pdf, href formula.

df.drop(['unnamed: 22', 'unnamed: 23', 'case number', 'original order', 'investigator or source', 'pdf', 'href formula'], axis=1)
```

```python
#Establece la columna Case Number como Ã­ndice del DataFrame.

df.set_index(['unnamed: 0'], inplace=True)
df.head()
```

```python
#Guarda el csv que no tiene esas columnas para seguir usÃ¡ndolo maÃ±ana.

df.to_csv('df_csv_modificado.csv')
```
