```python
conteo=df.shape[0]
```

```python
#쮺u치l es el porcentaje de valores nulos en cada columna?

nulos=df.isnull().sum()
porcentaje=nulos/conteo * 100
porcentaje
```

```python
#Sacad los valores 칰nicos y su frecuencia de las variables num칠ricas. Almacenad los resultados en un diccionario. Uno para cada variable.
dict_valores = {}  #Definimos un diccionario vac칤o para poder a침adir valores y key.

for col in df:
   unicos=df[col]. unique ()
   unicos_cont=df.value_counts(col)
   for k, v in zip(unicos, unicos_cont): 
      dict_valores[k] = v #Para cada key y value en el bucle for asignamos un valor a cada uno a partir del zip de las dos listas.

```

```python
#Cambiad el nombre de las columnas para que todas est칠n en min칰sculas

columns_minusculas = {col :  col.lower() for col in df.columns}
columns_minusculas
```

```python
df.rename(columns=columns_minusculas,inplace=True)
```

```python
df.head(1)
```

```python
# y reemplazad los espacios por _.

espacios_columns={col:col.replace(' ','_') for col in df.columns}
espacios_columns
```

```python

df.rename(columns=espacios_columns,inplace=True)
df

```

```python
#Eliminad las columnas de Unnamed: 22 y Unnamed: 23, Case Number.1, Case Number.2, original order, Investigator or Source, pdf, href formula.

df.drop(['unnamed:_22', 'unnamed:_23', 'case_number', 'original_order', 'investigator_or_source', 'pdf', 'href_formula'], axis=1, inplace=True)
```

```python
#Establece la columna Case Number como 칤ndice del DataFrame.

df.set_index(['unnamed:_0'], inplace=True)
df.head()
```
