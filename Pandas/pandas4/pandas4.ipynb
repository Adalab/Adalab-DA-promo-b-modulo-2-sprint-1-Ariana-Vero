```python
#8. Si extraemos los valores Ãºnicos de la columna species nos podemos dar cuenta que estÃ¡ un poco sucia 
# y extraer informacion Ãºtil de la misma se puede convertir en algo difÃ­cil. 
# En este ejercicio tendrÃ©is que extraer los ataques que han ocurrido por tiburÃ³n blanco 
# (lo encontraremos en la columna species como algo que contenga white o White). 
# Usad el mÃ©todo str.contains() para resolver este ejercicio.

df['species_'].unique()
```

```python
patron= 'white|White'

df_blanco=df[df["species_"].str.contains(patron, regex=True)]
df_blanco.shape
```

```python
#9. Extraed ahora los ataques por tiburÃ³n blanco o tigre. Utilizad el mÃ©todo str.contains() para resolver este ejercicio.

patron2= 'white|White|tiger|Tiger'

df_blanco_tigre=df[df["species_"].str.contains(patron2, regex=True)]
df_blanco_tigre.shape
```

```python
#10. Utilizando la columna de year cread una nueva columna llamada siglo que tiene siglo-XX o siglo-XXI.
#  Donde el aÃ±o sea menor que 2000 pondremos siglo-XX y el resto de los aÃ±os siglo-XXI.

df["century"] = np.where(df["year"] >2000,
                            "XX", "XXI")
df.head()
```
