```python
#¬øCu√°ntos ataques hubo por pa√≠s? Devuelve un DataFrame.
#üìå NOTA Ordena los resultados de mayor a menor. Esto todav√≠a no lo hemos visto, pero un DataFrame lo podemos ordenar con .sort_values(). Aqu√≠ ten√©is una explicaci√≥n de c√≥mo hacerlo.

ataques_pais = (df.groupby('country'))['country'].count().sort_values(ascending= False)
df_ataques = pd.DataFrame(ataques_pais)
df_ataques
```

```python
#¬øCu√°ntos ataques ha habido por a√±o? Devuelve un DataFrame. De nuevo ordena los resultados de mayor a menor.

ataques_a√±o = (df.groupby("year")['year']).count().sort_values(ascending= False)
df_ataques_a√±o = pd.DataFrame(ataques_a√±o)
df_ataques_a√±o
```

```python
#Cu√°ntos ataques hubo por a√±o y sexo? Devolved un DataFrame solo del a√±o 2000 en adelante.

ataques_a√±o_sexo = df.groupby(["year", 'sex_'])["case_number"].count().reset_index() #Ataques por a√±o y sexo 

ataques_mas_2000 = ataques_a√±o_sexo[ataques_a√±o_sexo['year']>2000] #Ataques por a√±o y sexo solo del 2000 en adelante.
ataques_mas_2000
```
