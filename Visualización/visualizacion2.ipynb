```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#Ayer creamos un par de gráficas para USA y España, donde visualizamos el número de ataques en cada país a lo largo de los años y el tipo de los ataques. Hoy tendréis que hacer dos gráficas con subplots:

df_usa=pd.read_csv('usa.csv')
df_spain=pd.read_csv('spain.csv')
```

```python
df_usa['type']
```

```python
df_usa.index
```

```python
#1. En la primera debéis mostrar los tipos ataques en USA y España, cada subplot corresponderá a un país. Debéis:
    #Poner nombre a los ejes de cada uno de los subplots
    #Poner un títulos a cada uno de los subplots
    #Poner el color de las líneas en negro.
    #Establecer el tamaño de la gráfica a 15 x 5
    #Ponerle marcadores a las gráficas
    #No incluir leyenda

fig,ax=plt.subplots(nrows=1, ncols=2, figsize=(15,5))

x1 = df_usa.index
y1 = df_usa['type']

x2 = df_spain.index
y2 = df_spain['type']

ax[0].plot(x1,y1, color = "black", marker = "x")
ax[0].set_title(label = "Ataques en USA")
ax[0].set_ylabel("Tipos de ataque")
ax[0].set_xlabel("Count")
ax[0].grid()


ax[1].plot(x2,y2, color = "black", marker = "v")
ax[1].set_title(label = "Ataques en España")
ax[1].set_ylabel("Tipos de ataque")
ax[1].set_xlabel("Count")
ax[1].grid()

plt.show()

plt.savefig('ataques_por_tipo.jpg')
```

```python
#2. En la segunda debéis mostrar los ataques por año para cada uno de los países.
    #De nuevo tendréis que ponerle nombre a los ejes y título a cada gráfica.
    #En este caso el subplot deberá tener dos filas y una columna
    #No incluir leyenda
    #Cambiar el color de las barras

fig,ax=plt.subplots(nrows=2, ncols=1, figsize=(20,20))

x1 = df_usa.index
y1 = df_usa['year']

x2 = df_spain.index
y2 = df_spain['year']

ax[0].bar(y1,x1, color = "g")
ax[0].set_title(label = "Ataques en USA por año")
ax[0].set_xlabel("Años")
ax[0].set_ylabel("Count")
ax[0].grid()


ax[1].bar(y2,x2 ,color = "r")
ax[1].set_title(label = "Ataques en España por año")
ax[1].set_xlabel("Años")
ax[1].set_ylabel("Count")
ax[1].grid()

plt.show()

plt.savefig('ataques_por_año.jpg')

```

```python
df_ataques=pd.concat((df_spain, df_usa),join='inner', ignore_index=False)
df_ataques.reset_index(inplace=True)
```

```python
#3. Cread un único gráfico donde juntéis las gráficas de los tipos de ataques en una sola. Debéis:
    #Poner leyenda a la gráfica
    #Cada línea debe ir en un color
    #Ponerle nombre a los ejes y a la gráfica.
    #Poner marcadores a las lineas.

fig=plt.subplots(figsize=(15,5))
fig=sns.histplot(x='type', data=df_ataques, hue='country', bins=8)

plt.savefig('ataques_por_tipo_y_pais.jpg')
```
