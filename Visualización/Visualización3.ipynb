```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
df=pd.read_csv('../Pandas/pandas5/merge.csv')
```

```python
df.head()
```

```python
#1. Usando la librería seaborn, crea una gráfica de barras que muestre el número de personas que han hecho cada actividad en cada año. Hacedlo solo desde el 2015. Os podrá salir una gráfica un poco fea, no os preocupeis, es lo que esperamos.

años2= df[df['year']>=2015]
años2.head(1)

plt.figure(figsize =(50,4))

fig = sns.histplot(x='activity', data = años2, hue='year')
fig.set(xlabel='Actividad', ylabel = 'Personas')
fig.set_title('Histograma de número de personas por actividad y año' )
plt.show()
```

En la gráfica de arriba podemos observar en el eje x las distintas actividades que se realizan, teniendo en cuenta el número de personas que realizan la actividad. Todo esto está estructurado dependiendo del año, siguiendo la leyenda de los colores. 

```python
#2. Usando la librería seaborn, crea una gráfica de barras que muestre el número de ataques que han ocurrido cada año en función del sexo. Hacedlo solo desde el 2015.

plt.figure(figsize =(10,4))

fig = sns.histplot(x='sex_', data = años2, hue='year')
fig.set(xlabel='Género', ylabel = 'Ataques')
fig.set_title('Histograma de número de ataques por género y año' )
plt.show()
```

En la gráfica de arriba podemos observar en el eje x los géneros que existen en el dataset, relacionándose con el número de ataques que han sucedido a lo largo de los años desde el 2015(leyenda de colres).

```python
#3. ¿Cuántos ataques hubo por año? Seleccionad solo desde el año 1500.
años= df[df['year']>=1500]

plt.figure()

fig = sns.histplot(x='year', data = años)
fig.set(xlabel='Año', ylabel = 'Número de ataques')
fig.set_title('Histograma de número de ataques por año')
plt.show()
```

En la gráfica de arriba podemos observar en el eje x los distintos años en los que ocurren los ataques(eje y), teniendo en cuenta el número de ataques que suceden.

```python
#4. Seleccionad solo los registros de USA, AUSTRALIA y SOUTH AFRICA. Muestra el número de ataques por país y año desde el año 1900.

df_paises = df[(df["country"] == "USA") | (df["country"] == "AUSTRALIA") | (df["country"] == "SOUTH AFRICA")]
```

```python
df_paises= df_paises[df_paises['year']>1900]
```

```python
plt.figure()

fig = sns.histplot(x='year', data = df_paises, hue='country')
fig.set(xlabel='Año', ylabel = 'Número de ataques')
fig.set_title('Histograma de número de ataques por año')
plt.show()
```

En la gráfica de arriba podemos observar en el eje x los distintos años en los que ocurren los ataques. En el eje y, observamos el número de ataques según el año y el país. Todo esto está estructurado dependiendo del país en el que ocurre, siguiendo la leyenda de los colores. 

```python
#5. Realizad un boxplot donde se muestre los principales estadísticos del número de ataques por país.

df_paises = df[(df["country"] == "USA") | (df["country"] == "AUSTRALIA") | (df["country"] == "SOUTH AFRICA")]
df_paises.head(1)
```

```python
plt.figure()

fig = sns.boxplot(x= 'country', y='year', data = df_paises, hue='country')
fig.set(ylabel='Año', xlabel = 'Países')
fig.set_title('Diagrama de caja de número de ataques por año y país')
plt.show()
```

En la gráfica de caja de arriba, podemos observar los diferentes principales estadísticos(este diagrama muestra estos conceptos de por sí) por cada uno de los países de nuestro dataset. La mayoría ocurren en torno a los 2000, en los tres países por igual.
